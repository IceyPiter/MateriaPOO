Dados os valores de um depósito fixo mensal e um montante desejado, crie uma classe para determinar quantos meses serão necessários para acumular o montante desejado, considerando juros mensais de 0,5%.

```java
import java.util.Scanner;

Scanner s = new Scanner(System.in);
System.out.println("Qual a quantia que deseja depositar");
double a = s.nextFloat();
System.out.println("Qual a quantia que desejaalcançar (Lucro 0,5% ao mês)");
double b = s.nextFloat();
double c = (0.5/100) * a;
    
int  i = 0;
while(a < b){
    i += 1;
    a += c;
}
System.out.print("Serão necessarios " + i + " meses para alcançar o objetivo");
```
